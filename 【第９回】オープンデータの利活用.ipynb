##student-mat.csv,student-por.csvの読み込み（Load files 'student-mat.csv'and 'student-por.csv'）

```python
import pandas as pd

student_data_math = pd.read_csv('student-mat.csv',sep=';')
student_data_por = pd.read_csv('student-por.csv',sep=';')
```

##ヒストグラム集計（Histogram aggregation）

```python
import matplotlib.pyplot as plt

plt.hist(student_data_math['absences'])

plt.xlabel('absences')
plt.ylabel('count')

plt.grid(True)
```

##平均、中央値、最頻値（Mean, median, mode）

```python
print('平均値： ',student_data_math['absences'].mean())
print('中央値： ',student_data_math['absences'].median())
print('最頻値： ',student_data_math['absences'].mode())
```

##分散と標準偏差（Variance and standard deviation）

```python
student_data_math['absences'].var()
```

```python
student_data_math['absences'].std()
```

```python
import numpy as np

np.sqrt(student_data_math['absences'].var())
```
